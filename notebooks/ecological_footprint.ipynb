# Import excel file and first cleanning

We'll clean the data a bit and keep only the columns of interest.

```python
xls = pd.ExcelFile('../raw_data/NFBA_2023.xlsx')
df = pd.read_excel(xls, 4).iloc[4:205, :]

df = df.drop(columns = ['Unnamed: 1','Unnamed: 2','Unnamed: 4',
                        'Unnamed: 9','Unnamed: 10','Unnamed: 11',
                        'Unnamed: 12','Unnamed: 13','Unnamed: 14',
                        'Unnamed: 15'])

df.columns = list(df.iloc[0,:])

df = df.drop(index = 4)

df = df.drop(df.columns[[-3, -4, -5]],axis = 1)

df.columns = [i.lower().strip().replace('\n',' ').strip().replace(' ','_') for i in df.columns]

df
```

```python
df.info()
```

```python
# We have a huge dataframe with all the data. Let's split it according to the information it stores:

df = df.reset_index(drop = True)

reg_fp = df.iloc[:14,:]  # Information about regions
coun_fp = df.iloc[17:199,:] # Information about countries

# Let's convert the columns we can to nummeric type:

def to_num(x):
    try:
        return float(x)
    except:
        return x

for i in range(df.shape[1]):
    reg_fp.iloc[:,i] = reg_fp.iloc[:,i].apply(to_num)
    coun_fp.iloc[:,i] = coun_fp.iloc[:,i].apply(to_num)

# And drop some columns

reg_fp = reg_fp.drop(columns = ['country','life_exectancy','per_capita_gdp'])

reg_fp

reg_fp.info() # Better!
```

```python
# Let's convert the datetime columns to string dates:

reg_fp['official__country_overshoot_day__(for_use_in_2024)'] = [i.date() for i in reg_fp['official__country_overshoot_day__(for_use_in_2024)']]
reg_fp['official__country_ecological_deficit_day__(for_use_in_2024)'] = [i.date() for i in reg_fp['official__country_ecological_deficit_day__(for_use_in_2024)']]
coun_fp['official__country_overshoot_day__(for_use_in_2024)'] = [i.date() for i in coun_fp['official__country_overshoot_day__(for_use_in_2024)']]
coun_fp['official__country_ecological_deficit_day__(for_use_in_2024)'] = [i.date() for i in coun_fp['official__country_ecological_deficit_day__(for_use_in_2024)']]

coun_fp['life_exectancy'][coun_fp.life_exectancy == ' '] = np.NaN
coun_fp['life_exectancy'] = coun_fp['life_exectancy'].fillna(coun_fp['life_exectancy'].median())
```

```python
# Now we will build the regions table will the adequate values. 
# First we save the income groups from the original table:

income_groups_fp = reg_fp.iloc[10:,1:].reset_index(drop = True)

income_groups_fp
```

```python
# Now to the regions:

coun_fp['region_id'] = coun_fp.country_id.apply(assign_region)

# coun_fp.region_id.unique()

world_fp = reg_fp.iloc[0,:].drop(columns = 'income_group')

reg_fp = reg_fp.reset_index(drop = True).iloc[:9,:].drop(columns = ['income_group',
                                                          'number_of_countries_required',
                                                          'official_country_overshoot_day',
                                                          'official_country_ecological_deficit_day'])

reg_fp.region = range(1,10)

reg_fp
```

```python
sub_coun_fp = coun_fp[['country_id',
                       'region_id',
                       'population_(millions)',
                       'cropland_footprint',
                       'grazing_footprint',
                       'forest_product_footprint',
                       'carbon_footprint',
                       'fish_footprint',
                       'total_ecological_footprint_(consumption)',
                       'cropland',
                       'grazing_land',
                       'forest_land',
                       'fishing_ground',
                       'built_up_land',
                       'total_biocapacity']]

for i in range(3,sub_coun_fp.shape[1]): # We multiply by the population to get absolute gha values per country
    sub_coun_fp.iloc[:,i] = sub_coun_fp.iloc[:,i]*sub_coun_fp['population_(millions)']
    
for i in range(2,reg_fp.shape[1]-2): # Now we sum for each region and divide by sum of its countries populations
    for j in range(1,10):
        reg_fp.iloc[j-1,i] = (sub_coun_fp[sub_coun_fp.region_id == j].iloc[:,i+1].sum())/\
        (sub_coun_fp[sub_coun_fp.region_id == j].iloc[:,2].sum())
        reg_fp.iloc[j-1,1] = sub_coun_fp[sub_coun_fp.region_id == j].iloc[:,2].sum()
        
# We recalculate this:

reg_fp['ecological_(deficit)_or_reserve'] = reg_fp['total_biocapacity'] \
                                            - reg_fp['total_ecological_footprint_(consumption)'] 

# And this:

reg_fp['number_of_earths_required'] = reg_fp['total_ecological_footprint_(consumption)']/world_fp['total_biocapacity']

reg_fp
```
