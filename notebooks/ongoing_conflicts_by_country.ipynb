```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
import sys

sys.path.append('../src')
from functions import *
```

# Web scraping

```python
main = 'https://wisevoter.com/country-rankings/countries-currently-at-war/'

soup = bs(req.get(main).content, 'html.parser')

countries = soup.find_all('h3', class_ = "component3-state-names")
state_bans = soup.find_all('div', class_ = "description-state-ban")
descriptions = soup.find_all('p', class_ = "description-state-text-content")

df = pd.DataFrame(data = [[i.text for i in countries],
                          [i.text.split('help')[1].split('Conflict')[0].strip() for i in state_bans],
                          [i.text.split('help')[-1].strip() for i in state_bans],
                          [i.text for i in descriptions]]).T

df.columns = ['country_name','conflict_type','conflict_casualties','conflict_description']

df
```

```python
# Let's correct the name and change it to the country_id from the countries.csv file:

df['country_name'] = df.country_name.apply(to_proper_country_name)

cc = pd.read_csv('../clean_data/countries.csv')

df['country_name'] = [cc[cc.country_name == i]['country_id'].iloc[0] for i in df.country_name]

df.columns = ['country_id', 'conflict_type', 'conflict_casualties', 'conflict_description']

df
```

```python
# Nice. Let's save the dataframe to upload it to the SQL database:

df.to_csv('../clean_data/ongoing_conflicts.csv', index = False)
```
